```
df3 = df2
df3["Content"] = ""
```

```
for i in range(0,len(df2)):
  new_text = df2["Text"][i].replace("\n","")
  df3["Text"][i] = new_text
  downloaded = trafilatura.fetch_url(df2["URL"][i])
  content = trafilatura.extract(downloaded)
  try:
    content = content.replace("\n","")
  except:
    content = ""
  df3["Content"][i] = content
  print("\r{}".format(i),end="")
```
