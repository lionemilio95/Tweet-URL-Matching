```
def isValid(url):
  try:
    result = requests.get(url, verify=False, timeout=0.1)
    soup = BeautifulSoup(result.content, 'html.parser')
    if "twimg" in soup.link['href']:
      return False
    else:
      return True
  except:
    return False
```

```
def isValid2(url):
  session = requests.Session()
  try:
    resp = session.head(url, allow_redirects=True)
    if 'twitter' in resp.url:
      return False
    else:
      return True
  except:
    return False
```

```
for i in range(40000,df.shape[0]):
  urls = df.URL[i]
  print("\r{}% completed {}/56026".format(i/df.shape[0] *100, i), end="")
  if i % 2000 == 0:
    df.to_csv("clean_datav4.csv")
  if type(urls) is str:
    l_url = urls.split()
    l_valid = []
    if len(l_url) == 0:
      continue

    for url in l_url:
      if isValid(url):
        l_valid.append(url)
    df.URL[i] = ';'.join(l_valid)

```
